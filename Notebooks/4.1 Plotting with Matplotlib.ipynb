# 4.1 Plotting with Matplotlib

There are several visualisation libraries for Python.  One of the most venerable is Matplotlib.  

## Getting started

We'll be using the pyplot part of matplotlib (many libraries have sublibraries like this), se we need to import this library.

The "%matplotlib inline" command puts each plot onto this notebook: if you don't use it, your plots will appear in popup windows.

```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
```

## Your first lineplot

```python
x = np.linspace(0, 10, 100)
y = np.sin(x)
plt.plot(x, y)
```

## Saving a copy of your plot

```python
fig = plt.figure()
plt.plot(x, y)
fig.savefig('my_figure.png')
```

## Scatterplots

```python
x = np.array([1,4,3,2,6,4,7,8])
y = np.array([3,5,4,3,7,6,4,9])
plt.scatter(x, y)
```

```python
x = np.array([1,4,3,2,6,4,7,8])
y = np.array([3,5,4,3,7,6,4,9])
fig, ax = plt.subplots()
plt.scatter(x, y)
ax.set_ylabel('This is the Y axis')
ax.set_xlabel('This is the X axis')
ax.set_title('This is the Title')
```

## Column charts

```python
x = np.arange(8)
y = np.array([1,4,3,2,6,4,7,8])
width = 0.7
plt.bar(x,y, width)
```

```python
fig, ax = plt.subplots()

x = np.arange(8)
y = np.array([1,4,3,2,6,4,7,8])
l = ('a','B','c','d','E','f','g','h')
width = 0.7
ax.bar(x,y, width, color='r')
ax.set_ylabel('This is the Y axis')
ax.set_xlabel('This is the X axis')
ax.set_xticks(x+width/2)
ax.set_xticklabels(l)
ax.set_title('This is the Title')

fig.savefig('red_bars.png')
```
