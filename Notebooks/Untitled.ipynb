```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0, 10, 100)
y = np.sin(x)
plt.plot(x, y)
```

```python
x = np.array([1,4,3,2,6,4,7,8])
y = np.array([3,5,4,3,7,6,4,9])
plt.scatter(x, y)
```

```python
fig, ax = plt.subplots()
plt.scatter(x, y)
ax.set_ylabel('This is the Y axis')
ax.set_xlabel('This is the X axis')
ax.set_title('This is the Title')
```
